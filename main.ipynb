### Outliers

```python
columns = st_data_2018_clean.columns

for i in range(1, len(columns)):
    fig = plt.figure(figsize=(10,5))
    sns.boxplot(x=columns[i], data=st_data_2018_clean)
```

```python
articles_max = st_data_2018_clean['Scientific and technical journal articles'].max()
```

```python
st_data_2018_clean[st_data_2018_clean['Scientific and technical journal articles'] == articles_max]
```

```python
st_data_2018_clean[st_data_2018_clean['Scientific and technical journal articles'] > 60000]
```

### Correlation and scatterplots

```python
st_data_2018_clean.corr(method='pearson', numeric_only=True).style.background_gradient(cmap='viridis')
```

```python
st_data_2018_clean.corr(method='spearman', numeric_only=True).style.background_gradient(cmap='viridis')
```

```python
st_data_2009_clean.corr(method='spearman', numeric_only=True).style.background_gradient(cmap='viridis')
```

```python
st_data_2018_clean.plot.scatter(x='Scientific and technical journal articles', y='Patent applications, residents')
```

```python
st_data_2018_nonUS = st_data_2018_clean[st_data_2018_clean['Country Name'] != 'United States']
st_data_2018_nonUS.plot.scatter(x='Scientific and technical journal articles', y='Patent applications, residents')
```
